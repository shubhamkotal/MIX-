# Types of methods : Regular, Class and Static

Regular and class takes instance as 1st argumnet, lets see how to change it and take class as 1st argument

```python
class Employee:

    raise_amt = 20 # defined once
    
    def __init__(self,first,last,pay): # initializer or constructor # self - default notation
        self.first = first  #
        self.last = last #  instances
        self.pay = pay
    def fullname(self): # method, uses instance as attribute as 1st priority
        return '{} {}'.format(self.first,self.last)
    def salary(self):
        self.pay = int(self.pay * self.raise_amt)
        
    @classmethod
    def set_raise_amt(cls,amount): # notation cls
        cls.raise_amt = amount
                    
emp1= Employee('Shubham','kotal',1234)
emp2=Employee('Mr.Shubham','Kotal',12323)

Employee.set_raise_amt(50) # once classmethod is defined , we can change it

print(Employee.raise_amt)
print(emp1.raise_amt)
```

```python
#5
# Python program to demonstrate  
# use of class method and static method. 
from datetime import date 
   
class Person: 
    def __init__(self, name, age): 
        self.name = name 
        self.age = age 
       
    # a class method to create a Person object by birth year. 
    @classmethod
    def fromBirthYear(cls, name, year): 
        return cls(name, date.today().year - year) 
       
    # a static method to check if a Person is adult or not. 
    @staticmethod
    def isAdult(age): 
        return age > 18
   
person1 = Person('mayank', 21) 
person2 = Person.fromBirthYear('mayank', 1996) 
   
print (person1.age) 
print (person2.age) 
   
# print the result 
print (Person.isAdult(22)) 
```
